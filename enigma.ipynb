```python
#tekst do zaszyfrowania bez spacji
tekst = 'tutajwpisztextktorychceszzaszyfrowac'

#pozycje wirnika od 1 do 26
I_r = 1
II_r = 1
III_r = 1
```

```python
def enigma(letter, rotor1_pos=1, rotor2_pos=1, rotor3_pos=1):
    import numpy as np
    import string
    alphabet = ["".join(i) for i in string.ascii_lowercase]

    #rotory
    I   =    ['E K M F L G D Q V Z N T O W Y H X U S P A I B R C J'.lower()]
    II   =   ['A J D K S I R U X B L H W T M C Q G Z N P Y F V O E'.lower()]
    III   =  ['B D F H J L C P R T X V Z N Y E I W G A K M U S Q O'.lower()]

    I = I[0].split()
    II = II[0].split()
    III = III[0].split()
    
    #obracanie rotorow
    I_rot = [I[(i + rotor1_pos-1) % len(I)]
                   for i, x in enumerate(I)]
    II_rot = [II[(i + rotor2_pos-1) % len(II)]
                   for i, x in enumerate(II)]
    III_rot = [III[(i + rotor3_pos-1) % len(III)]
                   for i, x in enumerate(III)]

    I_d = dict(zip(alphabet, I_rot))
    II_d = dict(zip(alphabet, II_rot))
    III_d = dict(zip(alphabet, III_rot))

    I_d_r = dict(zip(I_rot,alphabet))
    II_d_r = dict(zip(II_rot,alphabet))
    III_d_r = dict(zip(III_rot,alphabet))

    #beben odwracajacy
    B = ['AY  BR  CU  DH  EQ  FS  GL  IP  JX  KN  MO  TZ  VW'.lower()]
    B = B[0].split()
    B_1 = [i[0] for i in B]
    B_2 = [i[1] for i in B]
    B = dict(zip(B_1,B_2))
    B.update(dict(zip(B_2, B_1)))
    
    #kodowanie
    
    index1 = I_d[letter]
    index2 = II_d[index1]
    index3 = III_d[index2]

    #beben odwracajacy
    returned = B[index3]

    returned1 =  III_d_r[returned]
    returned2 =  II_d_r[returned1]
    returned3 =  I_d_r[returned2]
    
    return returned3
```

```python
def enigma_text(text, rotor1_pos=1, rotor2_pos=1, rotor3_pos=1):
    a = ["".join(i) for i in text]
    cryptotext = []
    for i in a:
        cryptotext.append(enigma(i, rotor1_pos = rotor1_pos, rotor2_pos=rotor2_pos , rotor3_pos=rotor3_pos))
    
    cryptotext = "".join(cryptotext)
    return cryptotext
```

```python
enigma_text(tekst, rotor1_pos = I_r, rotor2_pos=II_r , rotor3_pos=III_r)
```
